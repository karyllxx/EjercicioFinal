```python
import time  # Importa la librería `time` para medir el tiempo de ejecución del algoritmo.
#Alumna: Karina Miranda Avila

def is_valid(board, row, col, num):
    """
    Verifica si el número puede ser colocado en la posición (row, col).
    """
    # Verificar la fila: comprueba si `num` ya está presente en la fila `row`.
    if num in board[row]:
        return False

    # Verificar la columna: comprueba si `num` ya está presente en la columna `col`.
    if num in [board[i][col] for i in range(9)]:
        return False

    # Verificar el subcuadrante 3x3 que contiene la celda (row, col).
    # Calcula las coordenadas del inicio del subcuadrante.
    start_row, start_col = 3 * (row // 3), 3 * (col // 3)
    # Recorre el subcuadrante y verifica si `num` está presente.
    for i in range(start_row, start_row + 3):
        for j in range(start_col, start_col + 3):
            if board[i][j] == num:
                return False

    # Si `num` no viola ninguna restricción, es válido.
    return True


def solve_sudoku(board):
    """
    Resuelve el Sudoku utilizando backtracking.
    """
    # Recorre cada celda del tablero.
    for row in range(9):
        for col in range(9):
            # Si la celda está vacía (valor 0), intenta llenarla.
            if board[row][col] == 0:
                # Prueba los números del 1 al 9 en esta celda.
                for num in range(1, 10):
                    # Verifica si el número es válido en esta posición.
                    if is_valid(board, row, col, num):
                        # Asigna temporalmente el número a la celda.
                        board[row][col] = num
                        # Llama recursivamente para resolver el tablero con este cambio.
                        if solve_sudoku(board):
                            return True
                        # Si no es una solución válida, retrocede (backtracking).
                        board[row][col] = 0
                # Si ningún número es válido, no se puede resolver desde esta posición.
                return False
    # Si no quedan celdas vacías, el Sudoku está resuelto.
    return True


def print_board(board):
    """
    Imprime el tablero de Sudoku.
    """
    # Recorre cada fila del tablero.
    for row in board:
        # Imprime los números de la fila separados por espacios. Los ceros se representan como puntos.
        print(" ".join(str(num) if num != 0 else "." for num in row))


# Tablero de ejemplo: 0 representa las celdas vacías.
sudoku_board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9],
]

# Mide el tiempo de inicio de la ejecución.
start_time = time.time()

# Llama a la función para resolver el Sudoku.
if solve_sudoku(sudoku_board):
    # Si se resuelve con éxito, imprime el tablero resultante.
    print("Sudoku resuelto:")
    print_board(sudoku_board)
else:
    # Si no hay solución, lo indica.
    print("No se encontró solución.")

# Mide el tiempo final de ejecución.
end_time = time.time()

# Calcula y muestra el tiempo transcurrido.
print(f"Tiempo de ejecución: {end_time - start_time:.4f} segundos")
```
